```python
# -----------------------------
# Imports
# -----------------------------
from transformers import MarianMTModel, MarianTokenizer
from datasets import load_dataset
from sacrebleu.metrics import BLEU
```

```python

# -----------------------------
# Load Model
# -----------------------------
model_id = "Helsinki-NLP/opus-mt-en-fr"
tokenizer = MarianTokenizer.from_pretrained(model_id)
model = MarianMTModel.from_pretrained(model_id)
```

```python

# -----------------------------
# Load Dataset (auto-download)
# -----------------------------
data = load_dataset("opus100", "en-fr")

# Take smaller sample for testing
test_samples = data["test"].select(range(150))
```

```python
# -----------------------------
# Storage lists
# -----------------------------
eng_list = []
true_list = []
pred_list = []

refs = []
preds = []
```

```python
# -----------------------------
# Loop through samples
# -----------------------------
for item in test_samples:
    eng_text = item["translation"]["en"]
    fr_text = item["translation"]["fr"]
```

```python
# Tokenize
encoded = tokenizer(eng_text, return_tensors="pt", truncation=True)
```

```python
# Generate translation
generated = model.generate(
    **encoded,
    num_beams=4,
    max_length=120,
    early_stopping=True
)
```

```python
translated = tokenizer.decode(generated[0], skip_special_tokens=True)
```

```python
# Store results
eng_list.append(eng_text)
true_list.append(fr_text)
pred_list.append(translated)

refs.append(fr_text)
preds.append(translated)
```

```python
# -----------------------------
# Calculate BLEU Score
# -----------------------------
bleu_metric = BLEU()
bleu_score = bleu_metric.corpus_score(preds, [refs])
print("BLEU Score:", bleu_score.score)
```

```python
# -----------------------------
# Show Sample Outputs
# -----------------------------
print("\nSample Outputs\n")

# Use the smaller length to avoid IndexError
num_samples = min(5, len(eng_list))

for i in range(num_samples):
    print("English       :", eng_list[i])
    print("Actual French :", true_list[i])
    print("Model French  :", pred_list[i])
    print("-" * 60)
```
